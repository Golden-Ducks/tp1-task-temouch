```python
# Class 1: Sports & Athletics (Context: Winning/Medals)
doc1 = "The gold medal price is high effort"
doc2 = "Winning a gold medal needs a high jump"
doc3 = "Market for a gold medal is a trade of sweat"
doc4 = "The athlete will trade all for a gold medal"

# Class 2: Finance & Economy (Context: Market/Investment)
doc5 = "The gold bars price is high today"
doc6 = "Investing in gold bars needs a high rate"
doc7 = "Market for gold bars is a trade of money"
doc8 = "The bank will trade all for gold bars"
```

```python
import numpy as np
from sklearn.cluster import KMeans
import re
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text.split()
```

```python
def vectorize(docs, n):
  # Implement n-gram extraction and vectorization
    docs_ngrams = []

    for doc in docs:
        tokens = preprocess_text(doc)

        ngrams = []
        for i in range(len(tokens) - n + 1):
            ngrams.append(" ".join(tokens[i:i+n]))

        docs_ngrams.append(ngrams)
    vocabulaire = []
    for doc in docs_ngrams:
        for term in doc:
            if term not in vocabulaire:
                vocabulaire.append(term)

    X = np.zeros((len(docs), len(vocabulaire)))

    for i, doc in enumerate(docs_ngrams):
        for term in doc:
            j = vocabulaire.index(term)
            X[i][j] += 1

    return X
```

```python
# Training / Clustering

all_docs = [doc1, doc2, doc3, doc4, doc5, doc6, doc7, doc8]
```

```python
# 1-gram Experiment
X1 = vectorize(all_docs,1)
km1 = KMeans(n_clusters=2, random_state=42).fit(X1)

# 2-gram Experiment
X2 = vectorize(all_docs, 2)
km2 = KMeans(n_clusters=2, random_state=42).fit(X2)

print(f"1-gram clusters: {km1.labels_}")
print(f"2-gram clusters: {km2.labels_}")
```

```python
# compare accuracy and precision
from sklearn.metrics import accuracy_score

true_labels = [0,0,0,0,1,1,1,1]

def get_accuracy(true, pred):
    acc1 = accuracy_score(true, pred)
    acc2 = accuracy_score(true, [1 - x for x in pred])
    return max(acc1, acc2)

print("Accuracy 1-gram:", get_accuracy(true_labels, km1.labels_))
print("Accuracy 2-gram:", get_accuracy(true_labels, km2.labels_))
```
