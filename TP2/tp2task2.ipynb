```python
# Documents
D1 = "I love cats"
D2 = "Cats are chill"
D3 = "I am late"
```

```python
import re
```

```python
def preprocess(doc):
    doc = doc.lower()
    doc = re.sub(r"[^\w\s]", "", doc)
    return doc.split()
```

```python
def add_padding(tokens):
    # wrap tokens with start and end flags
    return ["<s>"] + tokens +["</s>"]
```

```python
def extract_windows(tokens, window_size=1):
    k = 2 * window_size + 1

    return [
        " ".join(tokens[i:i+k])
        for i in range(len(tokens) - k + 1)
    ]
```

```python
def build_vocab(all_windows):
    vocab = sorted(set(all_windows))
    return {w: i for i, w in enumerate(vocab)}
```

```python
def vectorize_doc(doc_windows, vocab):
    # return a binary vector: 1 if window in doc, 0 otherwise
    vector= [0] * len(vocab)
    for window in doc_windows:
        if window in vocab:
            vector[vocab[window]] = 1
    return vector
```

```python
# Run
all_docs = [D1, D2, D3]
```

```python
all_windows = []
docs_windows = []

for doc in all_docs:
    tokens=preprocess(doc)
    tokens= add_padding(tokens)
    windows = extract_windows(tokens, window_size=1)

    docs_windows.append(windows)
    all_windows.extend(windows)
```

```python
vocab = build_vocab(all_windows)
```

```python
vectors = []
for windows in docs_windows:
    vec = vectorize_doc(windows, vocab)
    vectors.append(vec)
```

```python
print("Vocabulaire:")
for k, v in vocab.items():
    print(v, ":", k)

print("\nVectors:")
for i, vec in enumerate(vectors):
    print(f"D{i+1}:", vec)
```
