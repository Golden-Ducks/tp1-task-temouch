```python
#task1
import re
stop_word={"are","is","am","will","the","an","from","you","this","i","not","to","in","on","for","and"}
```

```python
contractions_mot ={
    "won't ": "will not",
    "don't":"do not",
    "can't":"cannot"
}
```

```python
def contractions(text):
  for c,full in contractions_mot.items():
    text = text.replace(c,full)
  return text
```

```python
def preprocess(text):
  text = text.lower()

  text = contractions(text)

  text=re.sub(r"[^a-z0-9\s]","",text)
  tokens=text.split()
  tokens = [t for t in tokens if t not in stop_word]
  return tokens
```

```python
d1="Dates are hiiding from you today :'))"
d2="I won't ateend this Sunday evening 13 pm."
d3="Cats are watching angels outside :)"
```

```python
print("d1:",preprocess(d1))
print("d2:",preprocess(d2))
print("d3:",preprocess(d3))
```
