```python
import re
import numpy as np
from collections import Counter
from sklearn.cluster import KMeans
```

```python
stop_word ={    "are","is","am","the","a","an","and","to","of","in","on",
    "today","during"}
def preprocess(text):
  text=text.lower()
  text = re.sub(r"[^a-z\s]","",text)
  tokens=text.split()
  tokens =[t for t in tokens if t not in stop_word]
  return tokens
```

```python
def vect(doc,n=1):
  ngrams=[]
  processed=[]
  for d in doc:
    tokens=preprocess(d)
    ngram=[" ".join(tokens[i:i+n]) for i in range(len(tokens)-n+1)]
    processed.append(ngram)
    ngrams.extend(ngram)

  vocabulaire= sorted(set(ngrams))
  index={w:i for i,w in enumerate(vocabulaire)}

  print(vocabulaire)
  ind = np.zeros((len(doc),len(vocabulaire)))

  for i , d in enumerate(processed):
    counts = Counter(d)
    for w,c in counts.items():
      ind[i][index[w]]=c
  return ind

```

```python
doc = [
    "Muslims are praying and giving charity today",
    "People are fasting and praying during Ramadan",
    "The community gathers to pray and help the needy",

    "Customers are buying and giving feedback today",
    "People are shopping and giving reviews online",
    "The community gathers to buy and help new clients"
]
```

```python
X1 = vect(doc, 1)

km = KMeans(n_clusters=2, random_state=42)
km.fit(X1)

print("Clusters:", km.labels_)
```

```python
X2 = vect(doc, 2)
km2 = KMeans(n_clusters=2, random_state=42)
km2.fit(X2)

print("Bigram clusters:", km2.labels_)
```
